```python
import cv2
import matplotlib.pyplot as plt

# Charger image en niveaux de gris
img = cv2.imread("../data/raw/image1.png", cv2.IMREAD_GRAYSCALE)

# 1) Rduction du bruit
img_denoised = cv2.GaussianBlur(img, (5,5), 0)

# 2) Amlioration du contraste
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_contrast = clahe.apply(img_denoised)

# 3) Binarisation (texte noir / fond blanc)
_, img_binary = cv2.threshold(
    img_contrast, 0, 255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# Affichage
plt.figure(figsize=(12,4))

plt.subplot(1,4,1)
plt.title("Original")
plt.imshow(img, cmap="gray")
plt.axis("off")

plt.subplot(1,4,2)
plt.title("Denoised")
plt.imshow(img_denoised, cmap="gray")
plt.axis("off")

plt.subplot(1,4,3)
plt.title("Contraste")
plt.imshow(img_contrast, cmap="gray")
plt.axis("off")

plt.subplot(1,4,4)
plt.title("Binaire")
plt.imshow(img_binary, cmap="gray")
plt.axis("off")

plt.show()
```
