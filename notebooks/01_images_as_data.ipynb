```python
import cv2
import matplotlib.pyplot as plt

# Charger une image
img_color = cv2.imread("C:/Users/user/ocr-embeddings/data/raw/image1.png")

# Convertir en niveaux de gris
img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

# Infos importantes
print("Image couleur :", img_color.shape)
print("Image gris :", img_gray.shape)
print("Pixel exemple :", img_gray[30, 30])

# Affichage
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.title("Image couleur")
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(1,2,2)
plt.title("Image en niveaux de gris")
plt.imshow(img_gray, cmap="gray")
plt.axis("off")

plt.show()
```
